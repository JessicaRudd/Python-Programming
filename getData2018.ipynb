```python
import requests
# Import the json library
import json
import pandas as pd
from IPython.core.display import clear_output

from warnings import warn
from time import time
from time import sleep
from random import randint
```

```python
base = "https://games.crossfit.com/competitions/api/v1/competitions/"
start = base + "open/2018/leaderboards"
parameters = {"regional": 0}
```

```python
#region = 2 is worldwide
divisionList = [1,2,11]
```

```python
headers = {
    'User-Agent': 'James Chapman, example.com',
    'From': 'chapmanjlc@gmail.com'
}
```

```python
df = pd.DataFrame()

yearList = []
competitionTypeList = []

affilliateNameList= []
ageList = []
competitorIdList = []
competitorNameList = []
genderList = []
heightList = []
professionList = []

divisionIdList = []
professionList = []
regionIdList = []
regionNameList = [] ###
teamCaptainList = []
weightList = []

scoreOverallList= []
scoreOverallRank= []

scoreOneList = []
scoreTwoList = []
scoreThreeList = []
scoreFourList = []
scoreFiveList = []
scoreSixList = []

scoreOneDisplayList = []
scoreTwoDisplayList = []
scoreThreeDisplayList = []
scoreFourDisplayList = []
scoreFiveDisplayList = []
scoreSixDisplayList = []

scoreOneRankList = []
scoreTwoRankList = []
scoreThreeRankList = []
scoreFourRankList = []
scoreFiveRankList = []
scoreSixRankList = []

errorList = []
```

```python
start_time = time()
requestcount = 0
```

```python
for division in divisionList:

    parameters['division']=division
    parameters['page']=1
    
    requestcount =0
    
    # Make a get request with the parameters.
    response = requests.get(start, params=parameters,headers = headers)
    numPages = response.json()['pagination']['totalPages']

    for i in range(1,numPages+1):

        try:
            parameters['page']=i

            # Monitor the requests
            requestcount += 1
            elapsed_time = time() - start_time
            print('Request:{} of {} for Division {}; Frequency: {} requests/s'.format(requestcount, numPages, division, requestcount/elapsed_time))
            clear_output(wait = True)

            response = requests.get(start, params=parameters, headers = headers)

            # try again
            #if response.json()['statusCode'] == 408:
            #    response = requests.get(start, params=parameters, headers = headers)


            # Throw a warning for non-200 status codes
            if response.status_code != 200:
                warn('Request: {}; Status code: {}'.format(requests, response.status_code))

            data = response.json()
            table = pd.DataFrame.from_dict(data['leaderboardRows'])

            #entrant and scores are both html ad need converted to table
            entrantjson = json.loads(table['entrant'].to_json(orient='records'))
            entranttable = pd.DataFrame.from_dict(entrantjson)
            scorejson = json.loads(table['scores'].to_json(orient='records'))
            scoretable = pd.DataFrame.from_dict(scorejson)

            #gets record count
            records = len(entranttable)

            #creates list for types of competition
            yearList.append([data['competition']['year']] * records)
            competitionTypeList.append([data['competition']['competitionType']] * records)
            
            #creates entrant lists
            affilliateNameList.append(entranttable['affiliateName'])
            ageList.append(entranttable['age'])
            competitorIdList.append(entranttable['competitorId'])
            competitorNameList.append(entranttable['competitorName'])
            genderList.append(entranttable['gender'])
            heightList.append(entranttable['height'])

            divisionIdList.append(entranttable['divisionId'])
            professionList.append(entranttable['profession'])
            regionIdList.append(entranttable['regionId'])
            regionNameList.append(entranttable['regionName'])
            teamCaptainList.append(entranttable['teamCaptain'])
            weightList.append(entranttable['weight'])

            #this on return so doing it from records not multiplying out

            #score Lists
            scoreOverallList.append(json.loads(table['overallScore'].to_json(orient='records')))
            scoreOverallRank.append(json.loads(table['overallRank'].to_json(orient='records')))
            
            scoreOneList.append(pd.DataFrame.from_dict(json.loads(scoretable[0].to_json(orient='records')))['score'])
            scoreTwoList.append(pd.DataFrame.from_dict(json.loads(scoretable[1].to_json(orient='records')))['score'])
            scoreThreeList.append(pd.DataFrame.from_dict(json.loads(scoretable[2].to_json(orient='records')))['score'])
            scoreFourList.append(pd.DataFrame.from_dict(json.loads(scoretable[3].to_json(orient='records')))['score'])
            scoreFiveList.append(pd.DataFrame.from_dict(json.loads(scoretable[4].to_json(orient='records')))['score'])
            scoreSixList.append(pd.DataFrame.from_dict(json.loads(scoretable[5].to_json(orient='records')))['score'])

            scoreOneDisplayList.append(pd.DataFrame.from_dict(json.loads(scoretable[0].to_json(orient='records')))['scoreDisplay'])
            scoreTwoDisplayList.append(pd.DataFrame.from_dict(json.loads(scoretable[1].to_json(orient='records')))['scoreDisplay'])
            scoreThreeDisplayList.append(pd.DataFrame.from_dict(json.loads(scoretable[2].to_json(orient='records')))['scoreDisplay'])
            scoreFourDisplayList.append(pd.DataFrame.from_dict(json.loads(scoretable[3].to_json(orient='records')))['scoreDisplay'])
            scoreFiveDisplayList.append(pd.DataFrame.from_dict(json.loads(scoretable[4].to_json(orient='records')))['scoreDisplay'])
            scoreSixDisplayList.append(pd.DataFrame.from_dict(json.loads(scoretable[5].to_json(orient='records')))['scoreDisplay'])
               
            scoreOneRankList.append(pd.DataFrame.from_dict(json.loads(scoretable[0].to_json(orient='records')))['rank'])
            scoreTwoRankList.append(pd.DataFrame.from_dict(json.loads(scoretable[1].to_json(orient='records')))['rank'])
            scoreThreeRankList.append(pd.DataFrame.from_dict(json.loads(scoretable[2].to_json(orient='records')))['rank'])
            scoreFourRankList.append(pd.DataFrame.from_dict(json.loads(scoretable[3].to_json(orient='records')))['rank'])
            scoreFiveRankList.append(pd.DataFrame.from_dict(json.loads(scoretable[4].to_json(orient='records')))['rank'])
            scoreSixRankList.append(pd.DataFrame.from_dict(json.loads(scoretable[5].to_json(orient='records')))['rank'])


            # Pause the loop
            sleep(randint(8,15)/100)
        
        except:
            errorList.append([parameters]+[response.content])
```
